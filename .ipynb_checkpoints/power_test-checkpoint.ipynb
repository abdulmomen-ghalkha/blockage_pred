```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
```

```python
datadf = pd.read_csv("Deepsense_dataset.csv")
datadf.head()
```

```python
datadf.info()
```

```python
# Holds the processed power vectors
power_list = []
paths = []

# First pass: load and store, marking NaNs
for idx, row in tqdm(datadf.iterrows()):
    path = row.iloc[2]
    try:
        power = np.loadtxt(path)
        if len(power) != 64:
            print(f"Unexpected shape in {path}: {power.shape}")
            continue
        power_list.append(power)
        paths.append(path)
    except Exception as e:
        print(f"Error loading {path}: {e}")
        continue

# Convert to 2D array
power_array = np.stack(power_list)  # shape: (N, 64)

# Fill NaNs using forward fill, then backward fill along axis 0 (row-wise)
power_df = pd.DataFrame(power_array)
power_df_ffill = power_df.ffill(axis=0).bfill(axis=0)

# Add path column and name columns a0 to a63
power_df_ffill.insert(0, 'path', paths)
power_df_ffill.columns = ['path'] + [f'a{i}' for i in range(64)]
```

```python
from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm

# Helper function to load power vector from a path
def load_power(row):
    path = row.iloc[2]
    index = row.iloc[0]
    try:
        power = np.loadtxt(path)
        if len(power) != 64:
            return None  # Invalid length
        return (index, path, power)
    except Exception:
        return None

# Parallel loading
with ThreadPoolExecutor(max_workers=8) as executor:  # Adjust num threads if needed
    results = list(tqdm(executor.map(load_power, [row for i, row in datadf.iterrows()]), total= len(datadf)))

# Filter valid results
filtered = [r for r in results if r is not None]
indices = [r[0] for r in filtered]
paths = [r[1] for r in filtered]
power_array = np.stack([r[2] for r in filtered])  # shape: (N, 64)

# Convert to DataFrame and fill NaNs
power_df = pd.DataFrame(power_array)
power_df = power_df.ffill(axis=0).bfill(axis=0)

# Add path column
power_df.insert(0, 'index', indices)

power_df.insert(1, 'path', paths)

power_df.columns = ['index'] + ['path'] + [f'a{i}' for i in range(64)]
```

```python
power_df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
# Check the indexing
numeric_cols = power_df.columns[2:]
scaler = MinMaxScaler()
power_df[numeric_cols] = scaler.fit_transform(power_df[numeric_cols])
```

```python
def fix_and_save(row):
    path = row.iloc[1]

    power = np.iloc[2:]
    
    # Save to new file
    dir_path = os.path.dirname(path)
    filename = os.path.splitext(os.path.basename(path))[0]
    new_path = os.path.join(dir_path, f"{filename}_fixed.txt")
    np.savetxt(new_path, power)


    return new_path


# Parallel processing
with ThreadPoolExecutor(max_workers=8) as executor:
    new_paths = list(tqdm(executor.map(fix_and_save, [row for _, row in power_df.iterrows()]), total=len(power_df)))

# Optional: filter out failed paths
new_paths = [p for p in new_paths if p is not None]
```
